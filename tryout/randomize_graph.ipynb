```python
N = 15
G = nx.generators.random_graphs.binomial_graph(N, 0.05)
```

```python
nx.draw_circular(G) 
```

```python
out = nx.algorithms.swap.double_edge_swap(G, nswap=G.number_of_edges()) 
nx.draw_circular(G)
```
