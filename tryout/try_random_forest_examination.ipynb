```python
from sklearn.datasets import load_iris
iris = load_iris()

treename = 'tree_temp'

# Model (can also use single decision tree)
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10)

# Train
model.fit(iris.data, iris.target)
# Extract single tree
estimator = model.estimators_[5]

from sklearn.tree import export_graphviz
# Export as dot file
export_graphviz(estimator, out_file=f'{treename}.dot', feature_names=iris.feature_names,
                class_names=iris.target_names, rounded=True, proportion=False, precision=2,
                filled=True, max_depth=3) 

# Convert to png using system command (requires Graphviz)
from subprocess import call
call(['dot', '-Tpng', f'{treename}.dot', '-o', f'{treename}.png', '-Gdpi=600'])

# Display in jupyter notebook
from IPython.display import Image
Image(filename = f'{treename}.png') 
```
